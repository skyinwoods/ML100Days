2.    給定散點圖顏色

Note：要畫正弦曲線先設定一下x的取值範圍，從0到2π。要用到numpy模組。

numpy.pi 表示π
numpy.arange( 0 , 2π ,0.01)  從0到2π，以0.01步進

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)

plt.show()
```

散點圖
顏色由（X，Y）的角度給出。

注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)  #(平均值,概率分佈的標準差,size)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)  #arctan2 的用法
plt.scatter(X,Y, s =75, c=T, alpha=.5) 
#c=T: 因為 c有 sequence
#s 代表點點大小
#alpha 數字越大越透明

print(T)
#print 出來的浮點數可以直接查色碼

plt.xlim(-1.5,1.5) #取值範圍，最小值和最大值
plt.ylim(-1.5,1.5)

plt.show()
```
